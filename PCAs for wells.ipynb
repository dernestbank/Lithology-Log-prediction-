### Dimensionality reduction using PCA
Principal component analysis (PCA) is a technique that transforms a dataset of many features into principal components that "summarize" the variance that underlies the data Each principal component is calculated by finding the linear combination of features that maximizes variance, while also ensuring zero correlation with the previously calculated principal components

<em>pca and selecting K values</em>

### Step-2 : Computing Covariance Matrix

```python
def computeCovarianceMatrix(data):
    len_data = data.shape[0]
    covariance = data.T.dot(data)/len_data
    return covariance

cov_mat2 = computeCovarianceMatrix(DR_X2)   # or scaled_df.cov()
cov_mat2 = np.around(cov_mat2,3)
print("for well2 " ,cov_mat2)

cov_mat3 = computeCovarianceMatrix(DR_X3)   # or scaled_df.cov()
cov_mat3 = np.around(cov_mat3,3)
print("for well3 " ,cov_mat3)

cov_mat4 = computeCovarianceMatrix(DR_X4)   # or scaled_df.cov()
cov_mat4 = np.around(cov_mat4,3)
print("for well4 ", cov_mat4)
```

### Step-3 : Eigen Decomposition of Covariance Matrix

```python
# from numpy.linalg import eig
eig_values2, eig_vectors2 = np.linalg.eig(cov_mat2)
print('\nNumber of Eigenvectors : %s ' %len(eig_vectors2))
print('\nEigenvectors : \n\n%s' %eig_vectors2)
print('\nEigenvalues : \n\n%s' %eig_values2)


# from numpy.linalg import eig
eig_values3, eig_vectors3 = np.linalg.eig(cov_mat3)
print('\nNumber of Eigenvectors : %s ' %len(eig_vectors3))
print('\nEigenvectors : \n\n%s' %eig_vectors3)
print('\nEigenvalues : \n\n%s' %eig_values3)

# from numpy.linalg import eig
eig_values4, eig_vectors4 = np.linalg.eig(cov_mat4)
print('\nNumber of Eigenvectors : %s ' %len(eig_vectors4))
print('\nEigenvectors : \n\n%s' %eig_vectors4)
print('\nEigenvalues : \n\n%s' %eig_values4)
```

### Visualize above eigenvectors as a dataframe (or coefficient matrix)

```python
eigenvec_df2 = pd.DataFrame(eig_vectors2) 
eigenvec_df2.columns = ['eigvec_'+str(i+1) for i in range(0, len(eig_vectors2))]
eigenvec_df2.T

eigenvec_df3 = pd.DataFrame(eig_vectors3) 
eigenvec_df3.columns = ['eigvec_'+str(i+1) for i in range(0, len(eig_vectors3))]
eigenvec_df3.T

eigenvec_df4 = pd.DataFrame(eig_vectors4) 
eigenvec_df4.columns = ['eigvec_'+str(i+1) for i in range(0, len(eig_vectors4))]
eigenvec_df4.T
```

### Step-4 : Sort the eigenvalues in descending order

```python
indexes2 = eig_values2.argsort()[::-1]   
eig_values2 = eig_values2[indexes2] 
eig_vectors2 = eig_vectors2[:, indexes2]
sorted_eig_pairs2 = [(np.around(np.abs(eig_values2[i]),3), eig_vectors2[:,i]) for i in range(len(eig_values2))]
print('\nSorted Eigen-pairs (descending order) :\n')

sorted_eigenValues2 = []
for i in range(0, len(sorted_eig_pairs2)):
    print("eigenvec_{0} : {1}".format(str(i+1), str(sorted_eig_pairs2[i][0])))
    sorted_eigenValues2.append(sorted_eig_pairs2[i][0])
    
print("\nTotal Variance (sum of all eigenvalues) : ", round(sum(sorted_eigenValues2), 3))

indexes3 = eig_values3.argsort()[::-1]   
eig_values3 = eig_values3[indexes3] 
eig_vectors3 = eig_vectors3[:, indexes3]
sorted_eig_pairs3 = [(np.around(np.abs(eig_values3[i]),3), eig_vectors3[:,i]) for i in range(len(eig_values3))]
print('\nSorted Eigen-pairs (descending order) :\n')

sorted_eigenValues3 = []
for i in range(0, len(sorted_eig_pairs3)):
    print("eigenvec_{0} : {1}".format(str(i+1), str(sorted_eig_pairs3[i][0])))
    sorted_eigenValues3.append(sorted_eig_pairs3[i][0])
    
print("\nTotal Variance (sum of all eigenvalues) : ", round(sum(sorted_eigenValues3), 3))


indexes4 = eig_values4.argsort()[::-1]   
eig_values4 = eig_values4[indexes4] 
eig_vectors4 = eig_vectors4[:, indexes4]
sorted_eig_pairs4 = [(np.around(np.abs(eig_values4[i]),3), eig_vectors4[:,i]) for i in range(len(eig_values4))]
print('\nSorted Eigen-pairs (descending order) :\n')

sorted_eigenValues4 = []
for i in range(0, len(sorted_eig_pairs4)):
    print("eigenvec_{0} : {1}".format(str(i+1), str(sorted_eig_pairs4[i][0])))
    sorted_eigenValues4.append(sorted_eig_pairs4[i][0])
    
print("\nTotal Variance (sum of all eigenvalues) : ", round(sum(sorted_eigenValues4), 3))
```

## Step-5 : Decide the value of "K" (optimal no. of PCs)

```python
best_eig_pairs2 = [sorted_eig_pairs2[i][0] for i in range(0, len(sorted_eig_pairs2)) if sorted_eig_pairs2[i][0] > 1.0]
print("\nAccording to Kaiser's Stopping Rule :")
print("Number of PCs to be considered for PCA is : {}".format(str(len(best_eig_pairs2))))
print("Their Eigenvalues are : {}".format(str(best_eig_pairs2)))
```

```python
best_eig_pairs3 = [sorted_eig_pairs3[i][0] for i in range(0, len(sorted_eig_pairs3)) if sorted_eig_pairs3[i][0] > 1.0]
print("\nAccording to Kaiser's Stopping Rule :")
print("Number of PCs to be considered for PCA is : {}".format(str(len(best_eig_pairs3))))
print("Their Eigenvalues are : {}".format(str(best_eig_pairs3)))
```

```python
best_eig_pairs4 = [sorted_eig_pairs4[i][0] for i in range(0, len(sorted_eig_pairs4)) if sorted_eig_pairs4[i][0] > 1.0]
print("\nAccording to Kaiser's Stopping Rule :")
print("Number of PCs to be considered for PCA is : {}".format(str(len(best_eig_pairs4))))
print("Their Eigenvalues are : {}".format(str(best_eig_pairs4)))
```

## Step-6 : Projection matrix construction

```python
K = 3
projection_matrix2 = np.array([list(np.hstack(i[1].reshape(11,1))) for i in sorted_eig_pairs2[:]])
projection_matrix2 = projection_matrix2[:K]
print("\nProjection-Matrix Dimension ... ", projection_matrix2.shape)
projection_matrix2
```

```python
K = 2
projection_matrix3 = np.array([list(np.hstack(i[1].reshape(11,1))) for i in sorted_eig_pairs3[:]])
projection_matrix3 = projection_matrix3[:K]
print("\nProjection-Matrix Dimension ... ", projection_matrix3.shape)
projection_matrix3
```

```python
K = 3
projection_matrix4 = np.array([list(np.hstack(i[1].reshape(11,1))) for i in sorted_eig_pairs4[:]])
projection_matrix4 = projection_matrix4[:K]
print("\nProjection-Matrix Dimension ... ", projection_matrix4.shape)
projection_matrix4
```
